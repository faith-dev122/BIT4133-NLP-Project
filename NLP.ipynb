```python
# WEEK 2 - Fig 1: Bigram and Trigram Output
from nltk.util import ngrams
from nltk.tokenize import word_tokenize
from collections import Counter

text = "Natural Language Processing helps computers understand human language \
and build intelligent systems that can communicate with people naturally."

tokens = word_tokenize(text.lower())

# Generate bigrams
bigrams = list(ngrams(tokens, 2))
print("BIGRAMS (word pairs):")
print("=" * 50)
for bg in bigrams[:8]:
    print(f"  {bg}")

# Generate trigrams
trigrams = list(ngrams(tokens, 3))
print(f"\nTRIGRAMS (word triplets):")
print("=" * 50)
for tg in trigrams[:8]:
    print(f"  {tg}")

print(f"\nTotal bigrams: {len(bigrams)}")
print(f"Total trigrams: {len(trigrams)}")
```

```python
# WEEK 2 - Fig 2 & 3: POS Tagging Code and Output
import nltk

# Sentence 1 - Simple sentence
sentence1 = "The student writes excellent notes every day."
tokens1 = nltk.word_tokenize(sentence1)
tags1 = nltk.pos_tag(tokens1)

print("POS TAGGING EXAMPLE 1:")
print("=" * 50)
print(f"Sentence: {sentence1}")
print("\nWord --> POS Tag --> Meaning")
print("-" * 40)

tag_meanings = {
    'DT': 'Determiner', 'NN': 'Noun', 'NNS': 'Noun(plural)',
    'VBZ': 'Verb(3rd person)', 'VBP': 'Verb', 'VB': 'Verb(base)',
    'JJ': 'Adjective', 'RB': 'Adverb', 'IN': 'Preposition',
    'NNP': 'Proper Noun', 'PRP': 'Pronoun', 'CC': 'Conjunction',
    'CD': 'Number', 'MD': 'Modal', 'VBG': 'Verb(gerund)',
    'VBD': 'Verb(past)', 'VBN': 'Verb(past participle)'
}

for word, tag in tags1:
    meaning = tag_meanings.get(tag, tag)
    print(f"  {word:<15} {tag:<8} {meaning}")

# Sentence 2 - More complex
sentence2 = "John and Mary quickly built an intelligent chatbot using Python."
tokens2 = nltk.word_tokenize(sentence2)
tags2 = nltk.pos_tag(tokens2)

print(f"\nPOS TAGGING EXAMPLE 2:")
print("=" * 50)
print(f"Sentence: {sentence2}")
print("\nWord --> POS Tag --> Meaning")
print("-" * 40)
for word, tag in tags2:
    meaning = tag_meanings.get(tag, tag)
    print(f"  {word:<15} {tag:<8} {meaning}")
```

```python
# WEEK 2 - Fig 4 & 5: Language Prediction using N-grams
from collections import defaultdict
import random

# Build a simple bigram language model
text = """I love natural language processing.
I love machine learning and deep learning.
Natural language processing is part of artificial intelligence.
Machine learning helps computers learn from data.
Deep learning uses neural networks to solve problems."""

tokens = word_tokenize(text.lower())

# Build bigram model — for each word, what words follow it?
bigram_model = defaultdict(list)
for i in range(len(tokens) - 1):
    bigram_model[tokens[i]].append(tokens[i+1])

print("BIGRAM LANGUAGE MODEL — WORD PREDICTION")
print("=" * 55)
print("Given a word, what words commonly follow it?\n")

test_words = ['natural', 'machine', 'deep', 'language', 'i']
for word in test_words:
    if word in bigram_model:
        followers = list(set(bigram_model[word]))
        print(f"  After '{word}':  →  {followers}")

print("\nSENTENCE ANALYSIS — Word Frequency")
print("=" * 55)
freq = Counter(tokens)
top_words = [(w, c) for w, c in freq.most_common(10) if w.isalpha()]
for word, count in top_words:
    print(f"  '{word}': {count} times")
```

```python
# WEEK 3 - Fig 1 & 2: Sequence Labeling and HMM Concept
import nltk

print("SEQUENCE LABELING USING POS TAGS (HMM-based)")
print("=" * 55)
print("In NLTK, pos_tag() uses a Hidden Markov Model internally")
print("to predict the most likely tag sequence.\n")

sentences = [
    "John lives in Nairobi and works at a technology company.",
    "The quick brown fox jumps over the lazy dog.",
    "Artificial Intelligence is transforming the world rapidly.",
    "Barack Obama was born in Hawaii in the United States.",
    "The students submitted their Natural Language Processing assignments."
]

for i, sentence in enumerate(sentences, 1):
    tokens = nltk.word_tokenize(sentence)
    tags = nltk.pos_tag(tokens)
    print(f"Sentence {i}: {sentence}")
    print(f"Tagged sequence:")
    print(f"  {tags}")
    print()
```

```python
# WEEK 3 - Fig 3 & 4: Named Entity Recognition (NER)
import nltk
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

def extract_named_entities(sentence):
    tokens = nltk.word_tokenize(sentence)
    tags = nltk.pos_tag(tokens)
    chunks = nltk.ne_chunk(tags)

    entities = []
    for chunk in chunks:
        if hasattr(chunk, 'label'):
            entity_name = ' '.join(c[0] for c in chunk)
            entity_type = chunk.label()
            entities.append((entity_name, entity_type))
    return entities

print("NAMED ENTITY RECOGNITION (NER)")
print("=" * 55)
print("Identifying people, places, and organisations in text\n")

test_sentences = [
    "Barack Obama was the president of the United States of America.",
    "Apple Inc was founded by Steve Jobs in California.",
    "Nairobi is the capital city of Kenya in East Africa.",
    "Google and Microsoft are major technology companies.",
    "The United Nations headquarters is located in New York."
]

for sentence in test_sentences:
    entities = extract_named_entities(sentence)
    print(f"Text: {sentence}")
    if entities:
        print(f"Entities found:")
        for name, etype in entities:
            print(f"  → {name}  [{etype}]")
    else:
        print("  No entities detected")
    print()
```

```python
# WEEK 4 - Practical Task 1: Dependency Parsing
# Analysing grammatical structure of sentences
import spacy

nlp = spacy.load("en_core_web_sm")

print("DEPENDENCY PARSING USING spaCy")
print("=" * 60)
print("Dependency parsing identifies how words relate to each other")
print("in a sentence — which word depends on which.\n")

# The exact sentence from your lecturer's document
text = "The lecturer teaches Natural Language Processing."
doc  = nlp(text)

print(f"Sentence: '{text}'")
print()
print(f"{'Token':<20} {'Dependency':<20} {'Head Word':<20} {'POS Tag'}")
print("-" * 75)
for token in doc:
    print(f"  {token.text:<20} {token.dep_:<20} "
          f"{token.head.text:<20} {token.pos_}")

print("\nDEPENDENCY LABELS EXPLAINED:")
print("-" * 60)
dep_meanings = {
    'nsubj'  : 'Nominal subject — who is doing the action',
    'ROOT'   : 'Root verb — the main action of the sentence',
    'dobj'   : 'Direct object — what the action is done to',
    'det'    : 'Determiner — article like the/a',
    'compound': 'Compound noun — words forming a combined noun',
    'attr'   : 'Attribute — description linked to subject',
    'prep'   : 'Preposition — shows relationship',
    'pobj'   : 'Object of preposition',
    'amod'   : 'Adjective modifier',
    'advmod' : 'Adverb modifier'
}

for token in doc:
    meaning = dep_meanings.get(token.dep_, token.dep_)
    print(f"  '{token.text}' → [{token.dep_}] = {meaning}")
```

```python
# WEEK 4 Assignment 1: Dependency Analysis
import spacy

nlp = spacy.load("en_core_web_sm")

assignment_sentences = [
    "The administrator updated student records",
    "Machine learning improves language processing"
]

dep_meanings = {
    'nsubj'   : 'Subject — who does the action',
    'ROOT'    : 'Main verb — the core action',
    'dobj'    : 'Direct object — receives the action',
    'det'     : 'Determiner — the/a/an',
    'compound': 'Compound — words forming combined noun',
    'amod'    : 'Adjective modifier — describes a noun',
    'nmod'    : 'Noun modifier',
    'attr'    : 'Attribute',
    'prep'    : 'Preposition',
    'pobj'    : 'Object of preposition',
    'advmod'  : 'Adverb modifier'
}

for sentence in assignment_sentences:
    doc = nlp(sentence)
    print(f"SENTENCE: '{sentence}'")
    print("=" * 60)
    print(f"{'Token':<20} {'Dep Label':<15} {'Head Word':<20} {'Role'}")
    print("-" * 70)

    for token in doc:
        role = dep_meanings.get(token.dep_, token.dep_)
        print(f"  {token.text:<20} {token.dep_:<15} "
              f"{token.head.text:<20} {role}")

    # Identify subject, verb, object explicitly
    subject = [t.text for t in doc if t.dep_ == 'nsubj']
    verb    = [t.text for t in doc if t.dep_ == 'ROOT']
    obj     = [t.text for t in doc if t.dep_ in ('dobj','pobj')]

    print(f"\n  STRUCTURE SUMMARY:")
    print(f"  Subject (who)  : {subject}")
    print(f"  Verb   (action): {verb}")
    print(f"  Object (what)  : {obj}")
    print()
```

```python
# WEEK 4 Assignment 2: Semantic Similarity Comparison
# Compare similar AND unrelated sentence pairs
import spacy
import matplotlib.pyplot as plt

nlp = spacy.load("en_core_web_sm")

print("SEMANTIC SIMILARITY — SIMILAR vs UNRELATED SENTENCES")
print("=" * 65)

sentence_pairs = [
    # Similar pairs
    ("The student passed the examination",
     "The learner succeeded in the test",
     "SIMILAR PAIR 1"),

    ("The doctor treated the patient at the hospital",
     "The physician healed the sick person at the clinic",
     "SIMILAR PAIR 2"),

    ("Python is used for machine learning",
     "Programmers use Python to build AI models",
     "SIMILAR PAIR 3"),

    # Unrelated pairs
    ("The student passed the examination",
     "The chef cooked a delicious meal",
     "UNRELATED PAIR 1"),

    ("Machine learning improves language processing",
     "The elephant walked through the jungle",
     "UNRELATED PAIR 2"),

    ("Natural Language Processing is part of AI",
     "The football match ended in a draw",
     "UNRELATED PAIR 3"),
]

scores = []
labels = []

for s1_text, s2_text, label in sentence_pairs:
    s1    = nlp(s1_text)
    s2    = nlp(s2_text)
    score = s1.similarity(s2)
    scores.append(score)
    labels.append(label)

    print(f"{label}:")
    print(f"  S1: '{s1_text}'")
    print(f"  S2: '{s2_text}'")
    print(f"  Similarity: {score:.4f} — "
          f"{'HIGH similarity' if score > 0.9 else 'MEDIUM similarity' if score > 0.7 else 'LOW similarity'}")
    print()

# Bar chart
colors = ['#70AD47' if 'SIMILAR' in l else '#FF4444' for l in labels]
plt.figure(figsize=(12, 6))
bars = plt.bar(labels, scores, color=colors, alpha=0.85, edgecolor='white')
for bar, score in zip(bars, scores):
    plt.text(bar.get_x() + bar.get_width()/2,
             bar.get_height() + 0.01,
             f'{score:.3f}',
             ha='center', fontweight='bold', fontsize=10)
plt.axhline(y=0.9, color='green', linestyle='--',
            alpha=0.5, label='High similarity threshold')
plt.axhline(y=0.7, color='orange', linestyle='--',
            alpha=0.5, label='Medium similarity threshold')
plt.xticks(rotation=25, ha='right', fontsize=9)
plt.ylabel('Similarity Score (0.0 to 1.0)')
plt.ylim(0, 1.15)
plt.title('Semantic Similarity — Similar vs Unrelated Sentence Pairs',
          fontweight='bold', fontsize=13)
plt.legend()
plt.tight_layout()
plt.savefig('semantic_similarity_chart.png', dpi=150)
plt.show()
```

```python
# WEEK 5 Practical Task 1: Complete NLP Pipeline
# Integrating ALL concepts from Weeks 1-4 into one pipeline
import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import pos_tag
from nltk.util import ngrams
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('wordnet')

nlp_spacy   = spacy.load("en_core_web_sm")
lemmatizer  = WordNetLemmatizer()
stop_words  = set(stopwords.words('english'))

def full_nlp_pipeline(text):
    print("=" * 65)
    print(f"INPUT TEXT: '{text}'")
    print("=" * 65)

    # STEP 1: Tokenization
    tokens = word_tokenize(text)
    print(f"\nSTEP 1 — TOKENIZATION:")
    print(f"  {tokens}")

    # STEP 2: Stopword removal
    filtered = [w for w in tokens if w.lower() not in stop_words
                and w.isalpha()]
    print(f"\nSTEP 2 — STOPWORD REMOVAL:")
    print(f"  {filtered}")

    # STEP 3: Lemmatization
    lemmatized = [lemmatizer.lemmatize(w.lower()) for w in filtered]
    print(f"\nSTEP 3 — LEMMATIZATION:")
    print(f"  {lemmatized}")

    # STEP 4: POS Tagging
    pos_tags = pos_tag(filtered)
    print(f"\nSTEP 4 — POS TAGGING:")
    for word, tag in pos_tags:
        print(f"  {word:<15} → {tag}")

    # STEP 5: N-grams
    bigrams_list  = list(ngrams(lemmatized, 2))
    print(f"\nSTEP 5 — BIGRAMS:")
    print(f"  {bigrams_list}")

    # STEP 6: Dependency Parsing (spaCy)
    doc = nlp_spacy(text)
    print(f"\nSTEP 6 — DEPENDENCY PARSING (spaCy):")
    for token in doc:
        print(f"  {token.text:<15} dep={token.dep_:<12} "
              f"head={token.head.text}")

    # STEP 7: Named Entity Recognition
    print(f"\nSTEP 7 — NAMED ENTITY RECOGNITION:")
    if doc.ents:
        for ent in doc.ents:
            print(f"  '{ent.text}' → [{ent.label_}]")
    else:
        print("  No named entities found")

    print("\nPIPELINE COMPLETE")
    print("=" * 65)

# Test sentences
sentences = [
    "Natural Language Processing helps computers understand language.",
    "Barack Obama studied at Harvard University in the United States.",
    "Google and Microsoft are building powerful artificial intelligence systems."
]

for sentence in sentences:
    full_nlp_pipeline(sentence)
    print()
```

```python
# WEEK 5 Mini Project: Student Academic Assistant Chatbot
# BIT4133 Faith Wanja Gichure
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords

nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

# ---- Knowledge base ----
responses = {
    # Greetings
    'hello'       : "Hello Faith! Welcome to the BIT4133 Academic Assistant.",
    'hi'          : "Hi there! How can I help you today?",
    'hey'         : "Hey! What would you like to know?",
    'good morning': "Good morning! Ready to help with your NLP queries.",
    'good afternoon': "Good afternoon! How can I assist you?",

    # CAT and exams
    'cat'         : "CAT 1 will test Weeks 1-5: tokenization, POS tagging, N-grams, HMM, parsing and semantic similarity. Prepare your logbook!",
    'exam'        : "CAT 1 covers Weeks 1-5 content. Focus on the NLP pipeline, spaCy, and NLTK practicals.",
    'test'        : "The CAT 1 practical assessment checks your logbook and coding skills from Weeks 1-5.",
    'marks'       : "Marks come from: logbook tasks, extra work (creativity marks), CAT 1, and the online course for CAT 2.",
    'grade'       : "Your grade depends on logbook quality, CAT scores, and your online course certificate.",

    # Unit and registration
    'unit'        : "This unit is BIT4133 — Natural Language Processing with Deep Learning. It runs for 15 weeks.",
    'register'    : "For unit registration, visit the student portal or contact the registrar's office.",
    'registration': "Registration queries should be directed to the academic office or student portal.",
    'course'      : "BIT4133 covers NLP fundamentals, deep learning, and real-world AI applications.",
    'units'       : "Please visit the academic office for unit registration queries.",

    # Logbook
    'logbook'     : "Your logbook must have weekly screenshots, code outputs, and reflections. It is checked every week!",
    'github'      : "Push your notebooks to GitHub regularly. Your repo link should be on your logbook cover page.",
    'screenshot'  : "Screenshots should show both your code and the output. Include a figure caption for each one.",

    # Project
    'project'     : "Your main project is the SMS Spam Detection System — a text classification NLP project.",
    'spam'        : "The SMS Spam Classifier uses TF-IDF and Naive Bayes to detect spam with 97% accuracy.",
    'sentiment'   : "The Sentiment Analyser is your Mini Project 2 — trained on 50,000 IMDB movie reviews.",
    'dataset'     : "Projects use the UCI SMS Spam Collection (5,572 messages) and IMDB Reviews (50,000 reviews).",

    # NLP concepts
    'nlp'         : "NLP is Natural Language Processing — teaching computers to understand human language.",
    'tokenization': "Tokenization splits text into individual tokens (words or sentences).",
    'stemming'    : "Stemming reduces words to their root form — e.g. 'running' becomes 'run'.",
    'lemmatization': "Lemmatization returns the proper dictionary form — e.g. 'studies' becomes 'study'.",
    'pos'         : "POS tagging labels each word with its grammatical role: Noun, Verb, Adjective etc.",
    'ngram'       : "N-grams are sequences of N consecutive words. Bigrams=2 words, Trigrams=3 words.",
    'spacy'       : "spaCy is a modern NLP library used for dependency parsing and NER in Week 4.",
    'parsing'     : "Parsing analyses sentence structure — dependency parsing shows how words relate.",
    'ner'         : "Named Entity Recognition identifies people, places and organisations in text.",
    'hmm'         : "Hidden Markov Models predict sequences of hidden states — used in POS tagging.",
}

def preprocess_input(text):
    """
    Apply NLP preprocessing to user input
    before matching to responses.
    This shows the chatbot uses real NLP internally.
    """
    tokens     = word_tokenize(text.lower())
    filtered   = [w for w in tokens
                  if w.isalpha() and w not in stop_words]
    return filtered, tokens

def get_response(user_input):
    """
    Match user input to the best response.
    Uses token-level matching after NLP preprocessing.
    """
    tokens, raw_tokens = preprocess_input(user_input)
    input_lower        = user_input.lower()

    # Direct phrase matching first
    for key, response in responses.items():
        if key in input_lower:
            return response

    # Token-level matching
    for token in tokens:
        for key, response in responses.items():
            if token in key or key in token:
                return response

    return ("I don't have an answer for that yet. "
            "Try asking about: CAT, logbook, project, "
            "tokenization, spaCy, NER, or registration.")

# Simulate chatbot conversation
print("=" * 60)
print("   BIT4133 STUDENT ACADEMIC ASSISTANT CHATBOT")
print("   Natural Language Processing — Faith Wanja Gichure")
print("=" * 60)
print("Type your question and press Enter.")
print("Type 'bye' to exit.\n")

# Simulate a conversation for demonstration
# (Replace with input() for live interaction)
demo_questions = [
    "Hello",
    "When is the CAT exam?",
    "What is my project about?",
    "Tell me about tokenization",
    "What is spaCy used for?",
    "How do I update my logbook?",
    "What units should I register for?",
    "What is NER?",
    "Tell me about the dataset",
    "bye"
]

print("DEMO CONVERSATION:")
print("-" * 60)
for question in demo_questions:
    print(f"You : {question}")
    if 'bye' in question.lower():
        print("Bot : Goodbye Faith! Good luck with your NLP studies!")
        print("\nChatbot session ended.")
        break
    tokens, _ = preprocess_input(question)
    response  = get_response(question)
    print(f"Bot : {response}")
    print(f"      [NLP: tokens after preprocessing = {tokens}]")
    print()
```
