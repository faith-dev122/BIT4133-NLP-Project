```python
# BIT4133 - Week 1 Practical 1: Simple Tokenization

# Step 1: Install/import libraries
import nltk

# Download required NLTK data (only needed once per Colab session)
nltk.download('punkt')
nltk.download('punkt_tab')

from nltk.tokenize import word_tokenize, sent_tokenize

# Step 2: Define a sample text
text = """
Natural Language Processing is a fascinating field of Artificial Intelligence.
It helps computers understand and generate human language.
Applications include chatbots, translators, and sentiment analysis tools.
"""

# Step 3: Word Tokenization — splits text into individual words
word_tokens = word_tokenize(text)
print("=" * 50)
print("WORD TOKENIZATION")
print("=" * 50)
print(f"Total word tokens: {len(word_tokens)}")
print("Tokens:", word_tokens)

# Step 4: Sentence Tokenization — splits text into sentences
sent_tokens = sent_tokenize(text)
print("\n" + "=" * 50)
print("SENTENCE TOKENIZATION")
print("=" * 50)
print(f"Total sentence tokens: {len(sent_tokens)}")
for i, sentence in enumerate(sent_tokens, 1):
    print(f"Sentence {i}: {sentence.strip()}")

# Step 5: Show the difference between the original text and tokens
print("\n" + "=" * 50)
print("SUMMARY")
print("=" * 50)
print(f"Original text length (characters): {len(text)}")
print(f"Number of word tokens: {len(word_tokens)}")
print(f"Number of sentence tokens: {len(sent_tokens)}")
```

```python
# BIT4133 - Week 1 Practical 2: Advanced Tokenization + Graph

# Step 1: Import libraries
import nltk
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from collections import Counter

# Download required NLTK data
nltk.download('punkt')
nltk.download('punkt_tab')
nltk.download('stopwords')

# Step 2: Define a richer sample text
text = """
Natural Language Processing (NLP) is a subfield of linguistics, computer science,
and artificial intelligence concerned with the interactions between computers and
human language. NLP enables computers to process and analyze large amounts of
natural language data. Applications of NLP include machine translation, speech
recognition, sentiment analysis, chatbot development, and information extraction.
Modern NLP systems use deep learning models and neural networks to achieve
human-level language understanding. The rise of large language models has
transformed the field of NLP and artificial intelligence significantly.
"""

# Step 3: Tokenize and convert to lowercase
tokens = word_tokenize(text.lower())
print(f"Step 1 — Raw tokens count: {len(tokens)}")
print(f"Sample raw tokens: {tokens[:10]}")

# Step 4: Remove punctuation — keep only alphabetic tokens
alpha_tokens = [word for word in tokens if word.isalpha()]
print(f"\nStep 2 — After removing punctuation: {len(alpha_tokens)} tokens")

# Step 5: Remove stop words (e.g. "the", "is", "and")
stop_words = set(stopwords.words('english'))
filtered_tokens = [word for word in alpha_tokens if word not in stop_words]
print(f"Step 3 — After removing stop words: {len(filtered_tokens)} tokens")
print(f"Sample filtered tokens: {filtered_tokens[:10]}")

# Step 6: Count word frequencies
word_freq = Counter(filtered_tokens)
most_common_10 = word_freq.most_common(10)

print(f"\nTop 10 most frequent words:")
for word, count in most_common_10:
    print(f"  '{word}': {count}")

# Step 7: Plot the bar chart
words = [item[0] for item in most_common_10]
counts = [item[1] for item in most_common_10]

plt.figure(figsize=(12, 6))
bars = plt.bar(words, counts, color='steelblue', edgecolor='white', linewidth=0.8)

# Add value labels on top of each bar
for bar, count in zip(bars, counts):
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height() + 0.05,
        str(count),
        ha='center', va='bottom', fontsize=11, fontweight='bold'
    )

# Labels and formatting
plt.xlabel('Words', fontsize=13)
plt.ylabel('Frequency', fontsize=13)
plt.title('Top 10 Most Frequent Words After Tokenization & Preprocessing', fontsize=14, fontweight='bold')
plt.xticks(rotation=30, ha='right', fontsize=11)
plt.yticks(fontsize=11)
plt.ylim(0, max(counts) + 1)
plt.tight_layout()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.savefig('tokenization_frequency_graph.png', dpi=150)  # Saves the graph as a PNG
plt.show()

print("\nGraph saved as 'tokenization_frequency_graph.png'")
```
